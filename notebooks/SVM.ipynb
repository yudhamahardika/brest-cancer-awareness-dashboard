```python
import pandas as pd
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from collections import Counter

# Ambil teks dari kolom 'steaming data' dalam DataFrame
# Convert the 'full_text' column to strings before joining
df = ' '.join(data['text_for_tfidf'].astype(str).tolist())

# Hapus stopwords dan kata-kata tambahan yang ingin diabaikan
stopwords = set(STOPWORDS)
stopwords.update({'https', 'the', 'of', 'to', 'amp'})

# Buat Wordcloud
wc = WordCloud(stopwords=stopwords, background_color="black", max_words=500, width=800, height=400)
wc.generate(df)

# Hitung frekuensi kata-kata untuk bar chart
tokens = df.split()
word_counts = Counter(tokens)
top_words = word_counts.most_common(10)  # Ubah 10 menjadi jumlah kata yang ingin ditampilkan di bar chart
word, count = zip(*top_words)

# Tampilkan Wordcloud
plt.figure(figsize=(15, 6))

# Wordcloud
plt.subplot(1, 2, 1)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Wordcloud")

# Bar chart
plt.subplot(1, 2, 2)
plt.bar(word, count, color='skyblue')
plt.title("Frekuensi Kata")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()





```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(data['full_text'], data['Label'], test_size=0.2, random_state=42)
X_train1, X_test1, y_train1, y_test1 = train_test_split(data['full_text'], data['Label'], test_size=0.3, random_state=42)

print(f"Jumlah Data Latih: {len(X_train)}")
print(f"Jumlah Data Uji: {len(X_test)}")
print(f"Jumlah Data Latih1: {len(X_train1)}")
print(f"Jumlah Data Uji1: {len(X_test1)}")







```

```python
# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Ganti nilai NaN dengan string kosong di X_train
X_train = X_train.fillna('')
X_train1 = X_train1.fillna('')

# Transformasi data teks menjadi vektor TF-IDF
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
X_train_tfidf1 = tfidf_vectorizer.fit_transform(X_train1)
X_test_tfidf1 = tfidf_vectorizer.transform(X_test1)
```

```python
# Cetak nama fitur dari TF-IDF Vectorizer
feature_names = tfidf_vectorizer.get_feature_names_out()
# Cetak beberapa contoh hasil transformasi dari data pelatihan
print("Hasil transformasi dari data pelatihan:")
print(X_train_tfidf)
print(X_train_tfidf1)

# Cetak beberapa contoh fitur dari vektor TF-IDF
print("\nFitur dari vektor TF-IDF:")
print(feature_names)
```

```python
# Inisialisasi model SVM
from sklearn.svm import SVC

svm_model = SVC(kernel='linear', random_state=42)

# Melatih model dengan data pelatihan yang sudah di-vektorisasi
svm_model.fit(X_train_tfidf, y_train)
svm_model.fit(X_train_tfidf1, y_train1)

# Cetak parameter model
print("Parameter model SVM:")
print(f"Kernel: {svm_model.kernel}")
print(f"C: {svm_model.C}")
print(f"Intercept: {svm_model.intercept_}")
print(f"Support Vectors: {svm_model.support_vectors_}")
```

```python
# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Ganti nilai NaN dengan string kosong di X_train and X_train1
X_train = X_train.fillna('')
X_train1 = X_train1.fillna('')

# Fit the vectorizer on the combined training data (X_train + X_train1)
all_train_data = pd.concat([X_train, X_train1])
tfidf_vectorizer.fit(all_train_data)

# Transform all datasets using the fitted vectorizer
X_train_tfidf = tfidf_vectorizer.transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
X_train_tfidf1 = tfidf_vectorizer.transform(X_train1)
X_test_tfidf1 = tfidf_vectorizer.transform(X_test1)

# Inisialisasi model SVM
svm_model = SVC(kernel='linear', random_state=42)

# Melatih model dengan data pelatihan yang sudah di-vektorisasi (use either X_train_tfidf or X_train_tfidf1)
svm_model.fit(X_train_tfidf, y_train) # or svm_model.fit(X_train_tfidf1, y_train1)

# Menguji model pada data pengujian yang sesuai
y_pred = svm_model.predict(X_test_tfidf) # If you trained on X_train_tfidf
y_pred1 = svm_model.predict(X_test_tfidf1) # If you trained on X_train_tfidf1
```

```python
# Evaluasi performa model
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
accuracy1 = accuracy_score(y_test1, y_pred1)
print(f"Accuracy: {accuracy:2f}")
print(f"Accuracy1: {accuracy1:.2f}")
```

```python
# Output classification report
print(classification_report(y_test, y_pred))
#print(classification_report(y_test1, y_pred))  # This line causes the error
print(classification_report(y_test1, y_pred1)) # This line uses the proper prediction for y_test1
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(data['full_text'], data['Label'], test_size=0.2, random_state=42)
X_train1, X_test1, y_train1, y_test1 = train_test_split(data['full_text'], data['Label'], test_size=0.3, random_state=42)

print(f"Jumlah Data Latih: {len(X_train)}")
print(f"Jumlah Data Uji: {len(X_test)}")
print(f"Jumlah Data Latih1: {len(X_train1)}")
print(f"Jumlah Data Uji1: {len(X_test1)}")
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
model.fit(X_train_tfidf1, y_train1)
```

```python
predictions = model.predict(X_test_tfidf)
predictions1 = model.predict(X_test_tfidf1)

accuracy = accuracy_score(y_test, predictions)
accuracy1 = accuracy_score(y_test1, predictions1)
print("\nClassification Report:\n", classification_report(y_test, predictions))
print("\nClassification Report:\n", classification_report(y_test1, predictions1))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, predictions))
print("\nConfusion Matrix:\n", confusion_matrix(y_test1, predictions1))
print(f"Accuracy: {accuracy:2f}")
print(f"Accuracy1: {accuracy1:2f}")

```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
# Model SVM
svm_model = SVC(kernel='linear', random_state=42)

# Model Naive Bayes
nb_model = MultinomialNB()

# Metode cross-validation (StratifiedKFold)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Cross-validation untuk model SVM
svm_scores = cross_val_score(svm_model, X_train_tfidf, y_train, cv=cv, scoring='accuracy')

# Cross-validation untuk model Naive Bayes
nb_scores = cross_val_score(nb_model, X_train_tfidf, y_train, cv=cv, scoring='accuracy')

# Cetak hasil
print("SVM Cross-Validation Scores:", svm_scores)
print("SVM Average Accuracy:", svm_scores.mean())

print("Naive Bayes Cross-Validation Scores:", nb_scores)
print("Naive Bayes Average Accuracy:", nb_scores.mean())
```
